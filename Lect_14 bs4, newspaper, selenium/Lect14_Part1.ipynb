# Сбор данных из интернета

В библиотеке pandas есть функция `read_html()`, которая делает запрос к странице, находит в HTML-разметке теги `<table>`, преобразовывает эти данные в датафреймы и возвращает список таблиц.

```python
import pandas as pd
df = pd.read_html('https://ru.wikipedia.org/wiki/%D0%A6%D0%95%D0%A0%D0%9D')
```

```python
len(df)
```

```python
df[0] # подставьте индексы от 0 до 11, увидите все таблицы, которые были на странице
```

Если мы знаем, что ищем, среди всех таблиц на странице, то можем использовать аргумент `match`. Обратите внимание, что даже если на странице всего одна таблица, функция `read_html` вернет ее как список из одного элемента, поэтому в конце обязательно использовать индекс `[0]`.

```python
pd.read_html('https://ru.wikipedia.org/wiki/%D0%A6%D0%95%D0%A0%D0%9D', match='пожертвование')[0]
```
