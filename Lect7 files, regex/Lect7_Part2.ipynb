```python
import openpyxl
from docxtpl import DocxTemplate
from datetime import datetime
from docx2pdf import convert
```

**openpyxl**: библиотека для работы с файлами Excel формата .xlsx. Она позволяет читать, записывать и модифицировать такие файлы. https://openpyxl.readthedocs.io/en/stable/

**docxtpl**: библиотека Python, которая позволяет создавать документы Word на основе шаблонов. https://docxtpl.readthedocs.io/en/latest/

**datetime**: работа с датами и временем (входит в Стандартную библиотеку Python). https://docs.python.org/3/library/datetime.html

**docx2pdf**: библиотека Python, которая предоставляет функциональность для конвертации документов Word в формат PDF. https://pypi.org/project/docx2pdf/

```python
workbook = openpyxl.load_workbook('data.xlsx')
sheet = workbook.active
```

`workbook = openpyxl.load_workbook('data.xlsx')`: Здесь происходит загрузка файла Excel.

`sheet = workbook.active`: Здесь выбирается активный лист (т.е. лист, который отображается в данный момент при открытии файла Excel). Переменная sheet теперь представляет собой объект, с которым можно взаимодействовать для доступа к данным на листе.

`sheet.values` будет доставать с листа все значения.

```python
# нажмите на Tab после точки ниже, вы увидите список всех возможных операций с листом

sheet.
```

Все данные в этой таблицы – синтетические (сгенерированные с помощью библиотеки Faker).

```python
list_values = list(sheet.values)
print(list_values)
```

Чтобы этот код отработал в шаблоне, который вы заполняете обязательно должны быть "плейсхолдеры": {{название плейсхолдера}}. Откройте файл "template.docx" и посмотрите на его содержимое.

```python
doc = DocxTemplate("template.docx")

for info in list_values[1:]:
    doc.render({"name": info[0],
                "address": info[1],
                "phone": info[2],
                "email": info[3],
               "company": info[4],
               "date": datetime.now().strftime('%d/%m/%Y')})
    
    doc_name = f"data/Соглашение о намерении {info[0]} {info[4]}.docx"
    doc.save(doc_name)
```

Конвертируем все полученные файлы из .docx в .pdf. В Python это делается в одну строчку!

```python
convert('data','data_pdf')
```
