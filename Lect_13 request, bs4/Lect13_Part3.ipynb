### Cлучайный вопрос из [базы вопросов ЧГК](https://db.chgk.info/tour/mkm18_u)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://db.chgk.info/tour/mkm18_u' # возьмем конкретный турнир

page = requests.get(url)
page.encoding = 'utf-8'
soup = BeautifulSoup(page.text)

print(soup.prettify())
```

```python
questions = soup.find_all("div", {"class": "question"})
num_questions = len(questions)
num_questions
```

```python
questions[0].find_all('p')
```

```python
print(questions[0].find_all('p')[0].text.strip()) # на 0 индексе всегда вопрос
```

```python
print(questions[0].find_all('p')[1].text.strip()) # на 1 индексе всегда ответ
```

```python
print(questions[0].find_all('p')[2].text.strip()) # а дальше никакой систематизации
```

```python
print(questions[10].find_all('p')[2].text.strip())
```
